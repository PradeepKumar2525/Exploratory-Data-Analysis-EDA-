#Exploratory Data Analysis (EDA)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\PRADEEP\Downloads\train.csv")
```

```python
# Filling n/a with the median value
df['Age'].fillna(df['Age'].median(),inplace=True)
```

```python
#Fill missing Embarked with mode
df['Embarked'].fillna(df['Embarked'].mode()[0],inplace=True)
```

```python
# Drop Cabin due to high missing values
df.drop(columns=['Cabin'],inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.info()
```

```python
#Univariate Analysis

sns.histplot(df['Age'], kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
sns.histplot(df['Fare'], kde=True)
plt.title('Fare Distribution')
plt.show()
```

```python
# Count plots
sns.countplot(x='Sex', data=df)
plt.title('Gender Count')
plt.show()
```

```python
sns.countplot(x='Pclass', data=df)
plt.title('Passenger Class Count')
plt.show()
```

```python
#Bivariate Analysis

sns.barplot(x='Sex', y='Survived', data=df)
plt.title('Survival Rate by Gender')
plt.show()
```

```python
# Survival by class
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title('Survival Rate by Class')
plt.show()
```

```python
# Age distribution by survival
sns.boxplot(x='Survived',y='Age',data=df)
plt.title('Age Distribution by Survived')
plt.show()
```

```python
df = df.drop(columns=['Name', 'Ticket'])

df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
df['Embarked'] = df['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
```

```python
# Correlation heatmap
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
## 8. Insights
# - Females had much higher survival rate than males
# - 1st class passengers had higher survival rates
# - Younger passengers tended to survive more
# - Fare positively correlated with survival

# Summary
# This EDA revealed clear survival patterns based on gender, class, and age.
```
